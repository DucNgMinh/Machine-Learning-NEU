```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import decomposition
from sklearn.datasets import load_iris
import PCA
```

```python
df = load_iris().data
target = pd.DataFrame(load_iris().target, columns= ['Target'])
PCA = PCA.PCA(n_components = 2)
```

```python
reduced_iris = PCA.fit_transform(df)
reduced_iris = pd.DataFrame(reduced_iris, columns=['PCA1', 'PCA2'])
reduced_iris
```

```python
reduced_iris = pd.concat([reduced_iris, target], axis=1)
reduced_iris
```

```python
plt.figure(figsize= (6,6))
sns.scatterplot(x=reduced_iris['PCA1'], y= reduced_iris['PCA2'], hue=reduced_iris['Target'], palette= 'deep').set_title('PCA from scratch')
```

```python
var_explain, cumulative_var = PCA.explained_variance_raito()
```

```python
plt.plot(range(0,len(cumulative_var) + 1), np.concatenate(([0],cumulative_var)))
plt.ylabel('Cumulative explained variance')
plt.legend(loc='best')
plt.title('Explained variance')
plt.show()
```
