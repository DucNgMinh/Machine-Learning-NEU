```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

import K_means

import warnings
warnings.filterwarnings("ignore")
```

```python
n_samples = 150
n_components = 2

K = [2, 3, 4, 5]

X, y = datasets.make_multilabel_classification(n_samples=1500, n_features=2, random_state=0)

fig, ax = plt.subplots( 1,4, figsize=(20, 4))
for i, K in enumerate(K):
    k = K_means.K_means(K)
    k.fit_transform(X)
    x = k.data_and_labels()
    sns.scatterplot(x=x[:, 0], y=x[:, 1], hue=x[:, 2], palette='deep', ax= ax[i], legend= False, alpha= 0.5)

    centerpoint = k.get_centers()
    for point in centerpoint:
        ax[i].plot(point[0], point[1], '*',markersize=15)

plt.show()
```

```python
sse = []
for i in range(1,11):
    k = K_means.K_means(i)
    k.fit_transform(X)
    sse.append(k.inertia_())

plt.plot(range(1, 11), sse)
```
